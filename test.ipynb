```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.circuit import Parameter, ParameterVector
```

```python
n_qubits = 10
qr = QuantumRegister(n_qubits, 'qreg')
cr = ClassicalRegister(2, 'creg')

qc = QuantumCircuit(qr, cr)

#create parameter vector, length
phi = ParameterVector("phi", n_qubits)

for i in range(n_qubits):
    qc.ry(phi[i], qr[i])

qc.draw('mpl')
```

```python
import numpy as np

param_values = np.linspace(0, np.pi, n_qubits)
# bind the parameters after circuit to create a bound circuit

bc = qc.assign_parameters({phi: param_values}) #passs a dictionary of parameter values
bc.draw('mpl')
```
